```python
#Importing libraries
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Loading csv file
credit = pd.read_csv('BankChurners.csv')
#Displaying first 5 rows
credit.head()
```

## Finding the most appropiate graphs to convey information

```python
#First lets get rid of some columns that we don't want
credit = credit.iloc[:,0:-4]
credit.head()
```

```python
# Lets get more information on some columns to better understand this dataset
credit.info()
```

```python
# For the first objective, just a simple scatter plot should work
plt.scatter(credit.Customer_Age,credit.Months_on_book,color='blue',alpha=0.05)
plt.show()
```

### There's clearly correlation between age and months on book, just as expected, this lets us know that the data is trustworthy and realistic

```python
credit['Credit_Limit'].describe()
```

```python
# Once again, for the second objective, a simple scatter plot is all we need
sns.scatterplot(data=credit,x='Customer_Age',y='Credit_Limit')
plt.show()
```

### Now, this is interesting, even though there's a clear correlation between age and months on book, it seems there is no correlation whatsoever between age and credit limit, even young people can have bit credit limits, and older people have low credit limits

### That being said, having a lot of credit limit, does not mean that customers are spending efficiently, lets take a look at the total revolving balance compared with the credit limit

```python
credit['Total_Revolving_Bal'].describe()
```

```python
sns.scatterplot(data=credit,x='Total_Revolving_Bal',y='Credit_Limit')
plt.show()
```

While there's some outliers, there's a small tendency to have higher revolving balance with higher credit limit

```python
#Lets have a look at the same plot but separated by gender
sns.scatterplot(data=credit,x='Total_Revolving_Bal',y='Credit_Limit',hue='Gender')
plt.show()
```

Now theres a noticeable difference, males customers tend to have a higher revolving balance as well as credit limit, this means they are spending the most amount while keeping their balance up-to-date

Having a high credit limit, does not mean the customer uses that limit every month, they might use only some part of it, lets see

```python
credit.head()
```

```python
sns.lineplot(data=credit,x='Avg_Open_To_Buy',y='Credit_Limit',ci=None)
plt.show()
```

Unexpectedly, it seems theres a perfect correlation between the credit limit of customers and the average amount they are willing to spend

```python
#Grouping by education level
credit_education = credit.groupby('Education_Level').agg({'Avg_Open_To_Buy':'mean','Credit_Limit':'mean'}).round(2).reset_index()
credit_education = credit_education.iloc[0:6,:]
credit_education
```

```python
columns=list(credit_education.columns)
```

```python
plt.bar(x=credit_education[columns[0]],height=credit_education[columns[1]])
plt.xticks(rotation=30)
plt.show()
```
